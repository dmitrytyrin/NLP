# Preprocess

```python
morph = pymorphy2.MorphAnalyzer()
```

```python
with open('stopwords-ru.txt', 'r') as f:
    stop_words = f.read().split('\n')
```

```python
dict1 = {
    'usd':' доллар ',
    'eur':' евро ',
    'rur':' рубль ',
    '$':' доллар ',
    ' к. ':' корпус ',
    'кол-во':' количество ',
    ' т.к. ':' так как '
}
```

```python
dict2 = {
    ' тыс ':' тысяч ',
    ' д ':' дом ',
    ' ул ':' улица ',
    ' корп ':' корпус ',
    ' руб ':' рубль ',
    ' р ':' рубль ',
}
```

```python
def lemmatize_text(text):
    """Return lemmatized lower-cased string based on morph parser."""
    return ' '.join([morph.parse(w)[0].normal_form for w in text.split()])
```

```python
def clean_text(df, column_name, dictionary_before, dictionary_after):
    """Return dataframe with cleaned text in specified column."""
    # Replace from dictionary
    df.replace({column_name: dictionary_before}, inplace=True)
    
    # Replace english letters and numbers
    df[column_name] = df[column_name].str.replace('[A-Za-z0-9]+', ' ')
    # Replace special characters
    df[column_name] = df[column_name].str.replace('\W+',' ')

    df.replace({column_name: dictionary_after}, inplace=True)
    
    # Remove right and left whitespace
    df[column_name] = df[column_name].str.rstrip()
    df[column_name] = df[column_name].str.lstrip()

    # Lemmatization
    df[column_name] = df[column_name].apply(lemmatize_text)
    # Remove stop-words
    df[column_name] = df[column_name].apply(
        lambda x: ' '.join([item for item in x.split() if item not in stop_words]))
    
    return df
```

```python
df_all = clean_text(df_all, 'text_clean', dict1, dict2)
#df_all.head()
```

```python
len(df_all)
```

```python
# Tfidf vectorizer
tf = TfidfVectorizer(max_features=10000, lowercase=False).fit_transform(df_all.text_clean)
tf.shape
```

```python
# Dimensionality reduction for vizualization and baseline classification
dim_red = TruncatedSVD(random_state=1)
```

```python
matrix_2d = dim_red.fit_transform(tf)
matrix_2d
```

```python
matrix_2d.shape
```

```python
plt.figure(figsize=(10,10))
plt.scatter(matrix_2d[:claim_len,0], matrix_2d[:claim_len,1], c='r', alpha=0.01) # claims
plt.scatter(matrix_2d[claim_len:,0], matrix_2d[claim_len:,1], c='b', alpha=0.01) # conversions
plt.show()
```

We need to find blue points outside of the main blue cluster (on red cluster)

# Classification

```python
df_clf = pd.DataFrame(matrix_2d)
df_clf.columns = ['x','y']
df_clf['label'] = df_all.label
df_clf.head()
```

```python
clf = SVC(random_state=1)
```

```python
clf.fit(df_clf.iloc[:,:2].values, df_clf.iloc[:,2].values)
```

```python
df_clf['predict'] = clf.predict(df_clf.iloc[:,:2].values)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df_clf.x[df_clf.predict == 1], df_clf.y[df_clf.predict == 1], c='r', alpha=0.01)
plt.scatter(df_clf.x[df_clf.predict == 0], df_clf.y[df_clf.predict == 0], c='b', alpha=0.01)
plt.show()
```

# Join with initial dataframe

```python
# Claims in conversions
df_cross = df_clf.iloc[claim_len:,:][(df_clf.label == 0) & (df_clf.predict == 1)]
df_cross.head()
```

```python
df_final = df_all.iloc[:,:5].join(df_cross, how='inner')
#df_final.head()
```
